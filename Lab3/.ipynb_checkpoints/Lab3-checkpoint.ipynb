```python
df['Publishing date'] = pd.to_datetime(df['Publishing date'], format='mixed')

df['price'] = df['price'].replace(r'[\$,]', '', regex=True)

# Convert price to numeric
df['price'] = pd.to_numeric(df['price'])

# Display the new data types 
df.dtypes
```

```python
# Statistical summary
df.describe(include='all')
```

```python
sns.histplot(data=df, x='price', hue='form', bins=20, kde=True)
plt.title("Price Distribution by Book Format")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=20)
plt.title("Distribution of Price")
plt.show()
```

```python

df['price_cleaned'] = pd.to_numeric(df['price'].astype(str).str.replace('$', '', regex=False), errors='coerce')

genre_revenue = df.groupby('Genre')['price_cleaned'].sum().sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(10, 5))
genre_revenue.plot(kind='bar', color='teal')

# Adding labels and title
plt.title("Total Revenue by Genre")
plt.ylabel("Revenue")
plt.xlabel("Genre")

# 5. Rotate genre names for better readability
plt.xticks(rotation=45, ha='right')

# 6. Display the plot
plt.tight_layout()
plt.show()

# Display the numerical results
genre_revenue
```

```python
df['price_cleaned'] = pd.to_numeric(df['price'].astype(str).str.replace('$', '', regex=False), errors='coerce')

# Group by Book name and sum prices
product_revenue = df.groupby('Book name')['price_cleaned'].sum().sort_values(ascending=False)

# Plotting the Top 10 books by price/revenue
plt.figure(figsize=(10,5))
product_revenue.head(10).plot(kind='bar', color='orange')
plt.title("Revenue by Product (Top 10 Books)")
plt.ylabel("Revenue ($)")
plt.xticks(rotation=45, ha='right')
plt.show()

# Display numerical data
product_revenue.head(10)
```

```python
# 1. Clean the price column (removing '$' and converting to float)
df['price_cleaned'] = pd.to_numeric(df['price'].astype(str).str.replace('$', '', regex=False), errors='coerce')

# 2. Create the scatter plot
plt.figure(figsize=(8,5))
sns.scatterplot(x='reviews count', y='price_cleaned', data=df)

# 3. Add labels and title
plt.title("Reviews Count vs Price")
plt.xlabel("Reviews Count")
plt.ylabel("Price ($)")
plt.show()
```

```python
# 1. Clean the price column (ensuring it's numeric for correlation calculation)
df['price_cleaned'] = pd.to_numeric(df['price'].astype(str).str.replace('$', '', regex=False), errors='coerce')

# 2. Plot the Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df[['reviews count', 'price_cleaned']].corr(), annot=True, cmap='coolwarm')

# 3. Add title
plt.title("Correlation Matrix (Reviews vs Price)")
plt.show()
```

```python
# 1. Clean the price column
df['price_cleaned'] = pd.to_numeric(df['price'].astype(str).str.replace('$', '', regex=False), errors='coerce')

# 2. Convert 'Publishing date' to datetime and create the Month column
df['Publishing date'] = pd.to_datetime(df['Publishing date'], errors='coerce')
df['Month'] = df['Publishing date'].dt.to_period('M')

# 3. Group by Month and sum the prices
monthly_revenue = df.groupby('Month')['price_cleaned'].sum()

# 4. Plot the trend
plt.figure(figsize=(10,5))
monthly_revenue.plot(marker='o', color='b')
plt.title("Monthly Revenue Trend (Based on Publishing Date)")
plt.ylabel("Total Price ($)")
plt.xlabel("Month")
plt.grid(True)
plt.show()

# Display the monthly totals
monthly_revenue
```
